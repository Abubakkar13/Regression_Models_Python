## Catalog and Metal Stamping Specials Regression

### **Executive Summary**
Daily order and shipment data from the most recent 12 months were fitted to a regression line to determine any correlation between the variables.  
- Very little correlation was found between Catalog orders and Metal Stamping orders. 
- However, there is a slight positive correlation between Catalog shipments and Metal Stamping shipments.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["patch.force_edgecolor"] = True
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (8, 4)

# Load data
dataset = pd.read_csv('Catalog_MS_Regression_Daily.csv')

# Inpute outliers with mean
dataset['Metal_Stamping_Orders'] = dataset['Metal_Stamping_Orders'].replace(0,dataset['Metal_Stamping_Orders'].mean())
```

```python
# Joint plot of Catalog Orders vs Metal Stamping Orders
print('Catalog Orders vs Metal Stamping Orders:\nWith a correlation coefficient of 0.12, very little correlation exists.\nConfirmed by how dispersed the points appear away from the regression line.\nFindings shown below.')
sns.jointplot(x = 'Catalog_Orders', y='Metal_Stamping_Orders', 
              data = dataset, size = 7, ratio = 4, kind = 'reg', color = 'green')
plt.title('Catalog Orders vs Metal Stamping Orders')
plt.show()
```

```python
# Joint plot Catalog Shipments vs Metal Stamping Shipments
print('Catalog Shipments vs Metal Stamping Shipments:\nA weak positive correlation exists, as shown by the 0.32 Pearson coefficient.\nAnother indication of this relationship is how compact the points start to\ncluster around the regression line. Thus, an increase in Catalog shipments is\ntypically associated with a slight increase in Metal Stamping shipments.\nIt is worth noting that the strength of this relationship is relatively weak:\n.3 on a scale of -1 (perfect negative correlation) and 1 (perfect\npositive correlation). Results are below.')
sns.jointplot(x = 'Catalog_Shipments', y='Metal_Stamping_Shipments', 
              data = dataset, size = 7, ratio = 4, kind = 'reg', color = 'purple')
plt.title('Catalog Shipments vs Metal Stamping Shipments')
plt.show()
```
